```python
import pandas as pd
from datetime import datetime
import time
import numpy as np
from os import listdir
```

```python
file = "rota138FL"

df = pd.read_csv(f"../Raw/{file}.csv")

df
```

```python
df["DataIni"] = pd.to_datetime(df["DataIni"], unit="ms")
df["DataFim"] = pd.to_datetime(df["DataIni"], unit="ms")
df["HoraIni"] = pd.to_datetime(df["HoraIni"]).dt.time
df["HoraFim"] = pd.to_datetime(df["HoraFim"]).dt.time

df
```

```python
df.drop(['Sentido','Linha','NoVeículo','TotalGiros','KmPerc','DuraçãoViagem', 'Segundos', 'Unnamed: 0'], axis=1,inplace=True)
```

```python
df['DataInicio'] = df.apply(lambda x: datetime.combine(x['DataIni'],x['HoraIni']),axis=1)
df['DataFimm'] = df.apply(lambda x: datetime.combine(x['DataFim'],x['HoraFim']),axis=1)
```

```python
df.drop(['DataIni','HoraIni','DataFim','HoraFim'], axis=1,inplace=True)
```

```python
df['tempo_viagem'] = (df['DataFimm'] - df['DataInicio']).dt.seconds
```

```python
df['partidaTimeStamp'] = df['DataInicio'].values.astype(np.int64) // 10 ** 9
```

```python
df['dia_semana'] = df['DataInicio'].dt.dayofweek + 1
```

```python
df['tipo_dia'] = df['DataInicio'].dt.strftime('%Y-%m-%d') == '2019-09-07'
df['tipo_dia'].loc[df['tipo_dia'] == True] = 2
df['tipo_dia'].loc[df['tipo_dia'] == False] = 1
```

```python
df['hora_dia'] = df['DataInicio'].dt.hour
```

```python
df['turno_dia'] = df['hora_dia'].apply(lambda x: 1 if x < 5 else 4 if x > 17 else 2 if (x > 4 & x < 12) else 3 if (x > 11 & x < 18) else None)
```

```python
df['qtdDiasAno'] = df['DataInicio'].dt.dayofyear
```

```python
df.rename({"DataInicio": "data_partida"}, axis=1, inplace=True)
```

```python
df.to_csv(f"../Clean/{file}.csv",sep=',',index = False)
```
